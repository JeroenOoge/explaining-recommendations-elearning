```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
participants_df = pd.read_csv("../Datasets/data/participants.csv")
```

```python
colors = sns.color_palette("Paired", n_colors=4)
ax = sns.displot(
    data=participants_df,
    y="RG",
    linewidth=0,
    hue="Grade",
    hue_order=["9th grade", "10th grade", "11th grade", "12th grade"],
    multiple="stack",
    discrete=1,
    shrink=0.75,
    palette=colors,
    height=3,
    aspect=5/3
)
ax.set(xlabel="Number of Students", ylabel="")
plt.gcf().set_facecolor("white")
```

```python
import numpy as np
cross = pd.crosstab(participants_df["RG"], participants_df["Grade"])
table = pd.DataFrame(cross.to_records())
table.set_index(table["RG"], inplace=True)
table = table[["9th grade", "10th grade", "11th grade", "12th grade"]]
ax = table.plot.barh(stacked=True, color=["#a6cee3", "#1f78b4", "#b2df8a", "#33a02c"], width=.75)
ax.tick_params(axis="y", length=0)
ax.set_aspect("2")
ax.legend(loc="upper right", bbox_to_anchor=(1.35,1), edgecolor='none')
plt.gcf().set_facecolor("white")
plt.xlabel("Number of Participants")
plt.ylabel("")
plt.savefig("../Images/distribution_groups.pdf", format="pdf", bbox_inches="tight")
```
